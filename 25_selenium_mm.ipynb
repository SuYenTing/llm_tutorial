```python
import time
import random
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
```

```python
# Chrome設定
chrome_options = Options()
chrome_options.add_argument("--disable-blink-features=AutomationControlled")  # 移除自動化控制標記
# chrome_options.add_argument("start-maximized")  # 全螢幕
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36")

# 連線至MM 台灣總經成績單 頁面
driver = webdriver.Chrome(options=chrome_options)
driver.get("https://www.macromicro.me/macro/tw")

time.sleep(random.uniform(1.5, 3.0))  # 隨機等待
```

```python
# 取得當前頁面內容
html = driver.page_source
```

```python
print(html)
```

```python
# bs4解析網頁
soup = BeautifulSoup(html, "html.parser")
```

```python
# 尋找目標位置
collections = soup.find_all("div", class_="collection")
```

```python
# 印出指標資訊
for i in range(len(collections)):
    print(collections[i].find("a").text)
    print(collections[i].find("h6", class_="stat-name").text)
    print(collections[i].find("span", class_="val").text)
    print(collections[i].find("div", class_="desc limit-rows limit-rows-2").text)
    print('-'*30)
```

```python
# 關閉瀏覽器
driver.quit()
```
