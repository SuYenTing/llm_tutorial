```python
import requests
import json
```

```python
url = "https://mops.twse.com.tw/mops/api/t05st10_ifrs"

# # 查最新資訊
# payload = {
#     "companyId": "2330",
#     "dataType": "1",
#     "month": "",
#     "year": "",
#     "subsidiaryCompanyId": ""
# }

# 查歷史資訊
payload = {
    "companyId": "1101",
    "dataType": "2",
    "month": "6",
    "year": "103",
    "subsidiaryCompanyId": ""
}

# 設定 headers 避免被擋
headers = {
    "User-Agent": "Mozilla/5.0",
    "Content-Type": "application/json"
}

# 發送 POST 請求
response = requests.post(url, headers=headers, json=payload)

# 確認回應成功
if response.status_code == 200:
    data = response.json()
    print("原始 JSON：")
    print(json.dumps(data, ensure_ascii=False, indent=2))
else:
    print("錯誤，Status Code:", response.status_code)
```
