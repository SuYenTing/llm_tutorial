```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pprint import pprint
import time
import random
```

```python
# 搜尋新聞標題文字
searchTitle = '台泥'

# 搜尋頁面
page = 1

# 目標網址
url = f"https://money.udn.com/search/result/1001/{searchTitle}/{page}?search_type=title"

# 發送 GET 請求
response = requests.get(url)

# 確認回應成功
if response.status_code == 200:
    pprint(response.text)
else:
    print("錯誤，Status Code:", response.status_code)
```

```python
# 轉為bs4格式
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# 取得新聞連結
nodes = soup.find_all("div", class_="story__content")
links = list()
for node in nodes:
    links.append(node.find("a").get("href"))
```

```python
## 完整程式碼

# 搜尋新聞標題文字
searchTitle = '台泥'

links = list()
page = 0  # 搜尋起始頁面
while True:

    print(f'現在正在爬取頁面: {page}')

    # 目標網址
    url = f"https://money.udn.com/search/result/1001/{searchTitle}/{page}?search_type=title"

    # 發送 GET 請求
    response = requests.get(url)

    # 確認回應成功
    if response.status_code != 200:
        print("錯誤，Status Code:", response.status_code)
        break

    # 轉為bs4格式
    soup = BeautifulSoup(response.text, 'html.parser')

    # 取得新聞連結
    nodes = soup.find_all("div", class_="story__content")
    
    if len(nodes) == 0:
        break

    # 儲存新聞連結
    for node in nodes:
        links.append(node.find("a").get("href"))

    # 進入下一頁
    page = page+1

    # 暫停時間 避免爬取過快
    time.sleep(random.randint(1, 3))
```

```python
# 轉為bs4格式
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# 取得新聞標題
title = soup.find("h1", id="story_art_title")
title = title.get_text()
pprint(title)
```

```python
# 取得新聞時間
time = soup.find("time", class_="article-body__time")
time = time.get_text()
print(time)
```

```python
# 取得新聞內文
article = soup.find("section", id="article_body")
pprint(article)
```

```python
# 整理內文
text = article.get_text(separator="", strip=True)
pprint(text)
```
